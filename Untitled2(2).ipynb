```python

def make_hamook_base_adjacency_matrix():
    A = []
    for i in range(0,w*l):
        if i%l==0:    #start node of row
            row = np.zeros(w*l)
            row[i+1] = 1
            if math.floor(i/l) != w-1: #check the last row
                row[i+l] = 1
            A.append(row)
        elif (i%l) == l-1: #final node of each row
            row =  np.zeros(w*l)
            row[i-1] = 1
            if math.floor(i/l) != w-1:#check the last row
                row[i+l] = 1
            A.append(row)
        else:
            row =  np.zeros(w*l)
            row[i-1] = 1
            row[i+1] = 1
            A.append(row)
    A = np.array(A)
    return A


```

```python
def make_hamook_adjacency_matrix():
    adjacency_matrix = make_hamook_base_adjacency_matrix()
    hamook_plus_adjacency_matrix = np.copy(adjacency_matrix)
    hamook_adjacency_matrix = np.copy(adjacency_matrix)
    for i in range(0,w-1):
        for j in range(1,l-1):
            if(i%2==0):
                if(j%2 == 0):
                    hamook_adjacency_matrix[i*l+j][(i+1)*l+j] = 1
                    hamook_adjacency_matrix[(i+1)*l+j][i*l+j] = 1
                   
                    hamook_bars.append((i*l+j,(i+1)*l+j))
                else:
                    hamook_plus_adjacency_matrix[i*l+j][(i+1)*l+j] = 1
                    hamook_plus_adjacency_matrix[(i+1)*l+j][i*l+j] = 1
    
                    hamook_plus_bars.append((i*l+j,(i+1)*l+j))
            else:
                if(j%2 == 1):
                    hamook_adjacency_matrix[i*l+j][(i+1)*l+j] = 1
                    hamook_adjacency_matrix[(i+1)*l+j][i*l+j] = 1
                    
                    hamook_bars.append((i*l+j,(i+1)*l+j))
                else:
                    hamook_plus_adjacency_matrix[i*l+j][(i+1)*l+j] = 1
                    hamook_plus_adjacency_matrix[(i+1)*l+j][i*l+j] = 1
                    
                    hamook_plus_bars.append((i*l+j,(i+1)*l+j))
    return hamook_adjacency_matrix,hamook_plus_adjacency_matrix,hamook_bars,hamook_plus_bars
```

```python
def make_hamook_graph():
    hamook_adjacency_matrix,hamook_plus_adjacency_matrix,hamook_bars,hamook_plus_bars = make_hamook_adjacency_matrix()
    hamook_graph= nx.from_numpy_array(hamook_adjacency_matrix)
    hamook_plus_graph = nx.from_numpy_array(hamook_plus_adjacency_matrix)
    
    
    #color bars with red
    for u,v in hamook_graph.edges:
        if (u,v) in hamook_bars:
            hamook_graph[u][v]['color']='r'
        else:
            hamook_graph[u][v]['color']='blue'
    for u,v in hamook_plus_bars:
        if (u,v) in hamook_plus_bars:
            hamook_plus_graph[u][v]['color']='r'
        else:
            hamook_plus_graph[u][v]['color']='blue'
           
        
    
    hamook_colors = nx.get_edge_attributes(hamook_graph,'color').values()
    hamook_plus_colors = nx.get_edge_attributes(hamook_plus_graph,'color').values()
    
    nx.draw_spring(hamook_graph, with_labels = True,edge_color = hamook_colors)
    nx.draw_spring(hamook_plus_graph, with_labels = True,edge_color = hamook_plus_colors)
    return hamook_graph,hamook_plus_graph
    
```
